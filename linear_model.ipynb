```python
import torch
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_data =[1.0, 2.0, 3.0]
y_data =[2.0, 4.0, 6.0]
```

```python
# w= 1.0 #random guess, random value 1.0 
```

```python
# our model for the forward pass
def forward(x):
    return x*w

# loss function 
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y)**2
```

```python
w_list =[]
mse_list = []
for w  in np.arange(0.0 , 4.1, 0.1):
    print("w = ", round(w,1))
    l_sum = 0 
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = round(forward(x_val),3)
        l =round(loss(x_val, y_val),3)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print('MSE = ',round(l_sum/3,3))
    w_list.append(w)
    mse_list.append(round(l_sum/3, 3))
    
```

```python
plt.plot(w_list, mse_list)
plt.ylabel("Loss")
plt.xlabel("w")
plt.show()
```
